```python
import os
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úì Libraries imported successfully!")
```

```python
# Load data
print("Loading data...")
df = pd.read_csv('../03_Results/housing_data_with_features.csv')
df['Date'] = pd.to_datetime(df['Date'])

print(f"‚úì Data loaded: {len(df)} rows, {len(df.columns)} columns")
```

```python
# Define features and target (same as baseline)
feature_cols = [
    'Index_Lag1', 'Index_Lag3', 'Index_Lag12',
    'Mortgage_1Y', 'Mortgage_3Y', 'Mortgage_5Y', 'Prime_Rate',
    'Index_YoY_Change', 'Mortgage_5Y_Change',
    'Index_MA3', 'Index_MA12',
    'Month', 'Quarter',
    'Price_Above_MA3', 'MA3_MA12_Diff'
]

target_col = 'Index_MoM_Change'

print(f"Features: {len(feature_cols)}")
print(f"Target: {target_col}")
```

```python
# Prepare data (same as baseline)
print("Preparing data...")

df_clean = df[feature_cols + [target_col, 'Date', 'City', 'Year']].dropna()

SPLIT_YEAR = 2024
train_mask = df_clean['Year'] < SPLIT_YEAR
test_mask = df_clean['Year'] >= SPLIT_YEAR

X_train = df_clean.loc[train_mask, feature_cols]
y_train = df_clean.loc[train_mask, target_col]
X_test = df_clean.loc[test_mask, feature_cols]
y_test = df_clean.loc[test_mask, target_col]

print(f"Train: {len(X_train)} samples")
print(f"Test: {len(X_test)} samples")
```

# Part 2: Load Baseline Results

Load the baseline model results to compare.

```python
# Try to load baseline results (if available)
try:
    baseline_results = pd.read_csv('../03_Results/baseline_model_performance.csv')
    print("Baseline Model Performance:")
    print("="*60)
    display(baseline_results)
    
    baseline_test_r2 = baseline_results['Test_R2'].values[0]
    baseline_test_mape = baseline_results['Test_MAPE'].values[0]
    
    print(f"\n‚úì Baseline to beat:")
    print(f"  R¬≤ = {baseline_test_r2:.4f}")
    print(f"  MAPE = {baseline_test_mape:.2f}%")
    
except FileNotFoundError:
    print("‚ö† Baseline results not found. We'll still build advanced models!")
    baseline_test_r2 = None
    baseline_test_mape = None
```

# Part 3: Random Forest Model

**Why Random Forest?**
- Handles non-linear relationships (Linear Regression can't)
- Less prone to overfitting than single decision tree
- Provides feature importance
- No need to scale features

```python
print("="*80)
print("TRAINING RANDOM FOREST")
print("="*80)

# Initialize Random Forest
rf_model = RandomForestRegressor(
    n_estimators=100,      # Number of trees
    max_depth=15,          # Maximum depth of each tree
    min_samples_split=5,   # Minimum samples to split a node
    min_samples_leaf=2,    # Minimum samples in a leaf
    random_state=42,       # For reproducibility
    n_jobs=-1,             # Use all CPU cores
    verbose=1              # Show progress
)

print("\nModel parameters:")
print(f"  Trees: {rf_model.n_estimators}")
print(f"  Max depth: {rf_model.max_depth}")
print(f"  Min samples split: {rf_model.min_samples_split}")

print("\nTraining...")
rf_model.fit(X_train, y_train)
print("‚úì Training complete!")
```

```python
# Make predictions
print("Making predictions...")
rf_train_pred = rf_model.predict(X_train)
rf_test_pred = rf_model.predict(X_test)
print("‚úì Predictions generated")
```

```python
# Evaluate Random Forest
def calculate_metrics(y_true, y_pred, model_name=""):
    """Calculate all metrics"""
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    mask = y_true != 0
    mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    
    print(f"{model_name} Metrics:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R¬≤:   {r2:.4f}")
    print(f"  MAPE: {mape:.2f}%")
    
    return {'Model': model_name, 'RMSE': rmse, 'MAE': mae, 'R2': r2, 'MAPE': mape}

print("="*60)
print("RANDOM FOREST PERFORMANCE")
print("="*60)

rf_train_metrics = calculate_metrics(y_train, rf_train_pred, "RF Training")
print()
rf_test_metrics = calculate_metrics(y_test, rf_test_pred, "RF Test")

# Compare with baseline
if baseline_test_r2 is not None:
    print("\n" + "="*60)
    print("COMPARISON WITH BASELINE")
    print("="*60)
    r2_improvement = ((rf_test_metrics['R2'] - baseline_test_r2) / baseline_test_r2) * 100
    mape_improvement = ((baseline_test_mape - rf_test_metrics['MAPE']) / baseline_test_mape) * 100
    
    print(f"R¬≤ improvement: {r2_improvement:+.1f}%")
    print(f"MAPE improvement: {mape_improvement:+.1f}%")
    
    if rf_test_metrics['R2'] > baseline_test_r2:
        print("\n‚úì Random Forest BEATS baseline!")
    else:
        print("\n‚ö† Random Forest did not beat baseline")
```

# Part 4: XGBoost Model

**Why XGBoost?**
- Usually best performance for tabular data
- Handles complex patterns
- Built-in regularization prevents overfitting
- Very fast training

```python
print("="*80)
print("TRAINING XGBOOST")
print("="*80)

# Initialize XGBoost
xgb_model = xgb.XGBRegressor(
    n_estimators=100,       # Number of boosting rounds
    max_depth=6,            # Maximum depth of trees
    learning_rate=0.1,      # Step size for updates
    subsample=0.8,          # Fraction of samples for each tree
    colsample_bytree=0.8,   # Fraction of features for each tree
    random_state=42,
    n_jobs=-1,
    verbosity=1
)

print("\nModel parameters:")
print(f"  Estimators: {xgb_model.n_estimators}")
print(f"  Max depth: {xgb_model.max_depth}")
print(f"  Learning rate: {xgb_model.learning_rate}")

print("\nTraining...")
xgb_model.fit(X_train, y_train)
print("‚úì Training complete!")
```

```python
# Make predictions
print("Making predictions...")
xgb_train_pred = xgb_model.predict(X_train)
xgb_test_pred = xgb_model.predict(X_test)
print("‚úì Predictions generated")
```

```python
# Evaluate XGBoost
print("="*60)
print("XGBOOST PERFORMANCE")
print("="*60)

xgb_train_metrics = calculate_metrics(y_train, xgb_train_pred, "XGB Training")
print()
xgb_test_metrics = calculate_metrics(y_test, xgb_test_pred, "XGB Test")

# Compare with baseline
if baseline_test_r2 is not None:
    print("\n" + "="*60)
    print("COMPARISON WITH BASELINE")
    print("="*60)
    r2_improvement = ((xgb_test_metrics['R2'] - baseline_test_r2) / baseline_test_r2) * 100
    mape_improvement = ((baseline_test_mape - xgb_test_metrics['MAPE']) / baseline_test_mape) * 100
    
    print(f"R¬≤ improvement: {r2_improvement:+.1f}%")
    print(f"MAPE improvement: {mape_improvement:+.1f}%")
    
    if xgb_test_metrics['R2'] > baseline_test_r2:
        print("\n‚úì XGBoost BEATS baseline!")
    else:
        print("\n‚ö† XGBoost did not beat baseline")
```

# Part 10: City-Specific Analysis

**Why city-specific models?**
- Each city has unique market dynamics and volatility
- A single model treats all cities the same ‚Äî city-specific models can capture local patterns
- Helps identify which markets are most/least predictable

**Approach:**
1. Train a separate Random Forest model for each city
2. Compare performance across cities
3. Visualize with a heatmap

```python
print("="*80)
print("CITY-SPECIFIC MODEL ANALYSIS")
print("="*80)

cities = df_clean['City'].unique()
print(f"\nCities: {list(cities)}")

city_results = []

for city in sorted(cities):
    # Filter data for this city
    city_data = df_clean[df_clean['City'] == city]
    
    train_city = city_data[city_data['Year'] < SPLIT_YEAR]
    test_city  = city_data[city_data['Year'] >= SPLIT_YEAR]
    
    if len(test_city) < 5:
        print(f"  ‚ö† {city}: not enough test data, skipping")
        continue
    
    X_train_c = train_city[feature_cols]
    y_train_c = train_city[target_col]
    X_test_c  = test_city[feature_cols]
    y_test_c  = test_city[target_col]
    
    # Train Random Forest
    rf_city = RandomForestRegressor(
        n_estimators=100, max_depth=15,
        min_samples_split=5, min_samples_leaf=2,
        random_state=42, n_jobs=-1
    )
    rf_city.fit(X_train_c, y_train_c)
    pred_c = rf_city.predict(X_test_c)
    
    # Metrics
    r2   = r2_score(y_test_c, pred_c)
    rmse = np.sqrt(mean_squared_error(y_test_c, pred_c))
    mae  = mean_absolute_error(y_test_c, pred_c)
    mask = y_test_c != 0
    mape = np.mean(np.abs((y_test_c[mask] - pred_c[mask]) / y_test_c[mask])) * 100
    std  = y_test_c.std()
    
    city_results.append({
        'City': city,
        'R2': r2,
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape,
        'Volatility (œÉ)': round(std, 4)
    })
    print(f"  ‚úì {city:10} ‚Äî R¬≤: {r2:.4f}, RMSE: {rmse:.4f}, MAPE: {mape:.1f}%, œÉ: {std:.4f}")

city_df = pd.DataFrame(city_results).sort_values('R2', ascending=False)
print("\nCity-Specific Model Performance:")
print("="*60)
display(city_df)
```

```python
# Heatmap: City Performance vs Metrics
print("Generating city performance heatmap...")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Left: Heatmap of metrics per city ---
heatmap_data = city_df.set_index('City')[['R2', 'RMSE', 'MAE', 'Volatility (œÉ)']].copy()

# Normalize each column to 0-1 for color comparison
heatmap_norm = (heatmap_data - heatmap_data.min()) / (heatmap_data.max() - heatmap_data.min())

sns.heatmap(
    heatmap_norm,
    ax=axes[0],
    annot=heatmap_data.round(4),  # show actual values
    fmt='',
    cmap='RdYlGn',
    linewidths=0.5,
    cbar_kws={'label': 'Normalized Score'}
)
axes[0].set_title('City Performance Heatmap\n(color = normalized, values = actual)', 
                   fontsize=13, fontweight='bold')
axes[0].set_xlabel('')
axes[0].tick_params(axis='x', rotation=30)

# --- Right: R¬≤ bar chart per city ---
colors = ['#2ecc71' if r >= 0.85 else '#f39c12' if r >= 0.70 else '#e74c3c' 
          for r in city_df['R2']]
bars = axes[1].barh(city_df['City'], city_df['R2'], color=colors, alpha=0.8, edgecolor='black')

axes[1].set_xlabel('R¬≤ Score', fontsize=12)
axes[1].set_title('R¬≤ Score by City\n(green ‚â• 0.85, orange ‚â• 0.70, red < 0.70)', 
                   fontsize=13, fontweight='bold')
axes[1].axvline(x=0.85, color='green', linestyle='--', alpha=0.5, label='Excellent (0.85)')
axes[1].axvline(x=0.70, color='orange', linestyle='--', alpha=0.5, label='Good (0.70)')
axes[1].legend(fontsize=9)
axes[1].grid(alpha=0.3, axis='x')

for bar, val in zip(bars, city_df['R2']):
    axes[1].text(val + 0.005, bar.get_y() + bar.get_height()/2,
                f'{val:.4f}', va='center', fontsize=10)

plt.suptitle('City-Specific Random Forest Performance', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
os.makedirs('../03_Results/visualization', exist_ok=True)
plt.savefig('../03_Results/visualization/11_city_performance_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("‚úì Saved: 11_city_performance_heatmap.png")
```

```python
# ‚îÄ‚îÄ 2. City Value Comparison: Growth vs Volatility ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("Generating city value comparison chart...")

# Calculate stats per city from full dataset
city_stats = []
for city in sorted(df_clean['City'].unique()):
    city_data = df_clean[df_clean['City'] == city]
    
    total_growth = city_data[target_col].sum() * 100      # cumulative MoM sum as proxy
    volatility   = city_data[target_col].std()
    avg_monthly  = city_data[target_col].mean() * 100
    
    # Predictability from city_df if available
    pred_r2 = city_df[city_df['City'] == city]['R2'].values
    r2 = pred_r2[0] if len(pred_r2) > 0 else 0.5
    
    city_stats.append({
        'City': city,
        'Avg Monthly Growth (%)': round(avg_monthly, 4),
        'Volatility (œÉ)': round(volatility, 4),
        'Predictability (R¬≤)': round(r2, 4)
    })

stats_df = pd.DataFrame(city_stats)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# --- Left: Bubble chart ‚Äî Growth vs Volatility (bubble = R¬≤) ---
ax = axes[0]
for _, row in stats_df.iterrows():
    city = row['City']
    color = city_colors.get(city, '#555')
    size = row['Predictability (R¬≤)'] * 3000
    ax.scatter(row['Volatility (œÉ)'], row['Avg Monthly Growth (%)'],
               s=size, color=color, alpha=0.7, edgecolors='black', linewidth=1.5)
    ax.annotate(city, 
                (row['Volatility (œÉ)'], row['Avg Monthly Growth (%)']),
                textcoords='offset points', xytext=(8, 5), fontsize=10, fontweight='bold')

ax.set_xlabel('Volatility (œÉ) ‚Äî lower is more stable', fontsize=11)
ax.set_ylabel('Avg Monthly Growth (%)', fontsize=11)
ax.set_title('Growth vs Volatility\n(bubble size = predictability R¬≤)', 
             fontsize=12, fontweight='bold')
ax.grid(alpha=0.3)
ax.axhline(y=stats_df['Avg Monthly Growth (%)'].mean(), 
           color='gray', linestyle='--', alpha=0.5, label='Avg growth')
ax.axvline(x=stats_df['Volatility (œÉ)'].mean(), 
           color='gray', linestyle=':', alpha=0.5, label='Avg volatility')
ax.legend(fontsize=9)

# Add quadrant labels
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xmid = stats_df['Volatility (œÉ)'].mean()
ymid = stats_df['Avg Monthly Growth (%)'].mean()
ax.text(xlim[0]*1.02, ymid + (ylim[1]-ymid)*0.7, 'üèÜ Best\n(High growth,\nLow risk)', 
        fontsize=8, color='green', alpha=0.7)
ax.text(xlim[0]*1.02 + (xlim[1]-xlim[0])*0.55, ymid + (ylim[1]-ymid)*0.7, '‚ö†Ô∏è Risky\n(High growth,\nHigh risk)', 
        fontsize=8, color='orange', alpha=0.7)

# --- Right: Ranking table as bar chart ---
ax2 = axes[1]

# Score: normalize growth (higher better) - normalize volatility (lower better) + R2
stats_df['Score'] = (
    (stats_df['Avg Monthly Growth (%)'] - stats_df['Avg Monthly Growth (%)'].min()) /
    (stats_df['Avg Monthly Growth (%)'].max() - stats_df['Avg Monthly Growth (%)'].min()) * 0.4 +
    (1 - (stats_df['Volatility (œÉ)'] - stats_df['Volatility (œÉ)'].min()) /
    (stats_df['Volatility (œÉ)'].max() - stats_df['Volatility (œÉ)'].min())) * 0.3 +
    stats_df['Predictability (R¬≤)'] * 0.3
)
stats_df = stats_df.sort_values('Score', ascending=True)

bar_colors = [city_colors.get(c, '#555') for c in stats_df['City']]
bars = ax2.barh(stats_df['City'], stats_df['Score'], color=bar_colors, alpha=0.8, edgecolor='black')

ax2.set_xlabel('Overall Score\n(40% growth + 30% stability + 30% predictability)', fontsize=10)
ax2.set_title('City Investment Score Ranking\n(higher = better to buy)', 
              fontsize=12, fontweight='bold')
ax2.grid(alpha=0.3, axis='x')

for bar, val in zip(bars, stats_df['Score']):
    ax2.text(val + 0.005, bar.get_y() + bar.get_height()/2,
            f'{val:.3f}', va='center', fontsize=10, fontweight='bold')

# Medal labels
medals = ['ü•á', 'ü•à', 'ü•â']
for i, (bar, city) in enumerate(zip(reversed(list(bars)), reversed(list(stats_df['City'])))):
    if i < 3:
        ax2.text(0.002, bar.get_y() + bar.get_height()/2,
                medals[i], va='center', fontsize=12)

plt.suptitle('Where Should You Buy? City Value Comparison', 
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../03_Results/visualization/13_city_value_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("‚úì Saved: 13_city_value_comparison.png")

print("\n" + "="*60)
print("üí° INVESTMENT SUMMARY")
print("="*60)
top_city = stats_df.iloc[-1]
print(f"\nüèÜ Best city to buy: {top_city['City']}")
print(f"   Growth: {top_city['Avg Monthly Growth (%)']:+.4f}% avg/month")
print(f"   Volatility: {top_city['Volatility (œÉ)']:.4f} (stability)")
print(f"   Predictability R¬≤: {top_city['Predictability (R¬≤)']:.4f}")
print(f"\nüìä Full ranking:")
for _, row in stats_df.sort_values('Score', ascending=False).iterrows():
    print(f"   {row['City']:12} Score: {row['Score']:.3f}")
```
