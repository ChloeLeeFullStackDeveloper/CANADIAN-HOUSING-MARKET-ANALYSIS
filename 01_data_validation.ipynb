# 01. Data Validation

## Purpose
Validate data coverage, frequency, and basic quality checks.

```python
import pandas as pd

# Load datasets
rates_df = pd.read_csv("data/Daily Rates(in).csv")
prices_df = pd.read_csv("data/Monthly Home Price and Index by type and city(in).csv")

# Parse dates
rates_df['Date'] = pd.to_datetime(rates_df['Date'])
prices_df['Date'] = pd.to_datetime(prices_df['Date'])

# Basic info
print("Rates Data Shape:", rates_df.shape)
print("Prices Data Shape:", prices_df.shape)

# Check date ranges
print("Rates date range:", rates_df['Date'].min(), "→", rates_df['Date'].max())
print("Prices date range:", prices_df['Date'].min(), "→", prices_df['Date'].max())

# Check missing values
print("Rates missing values:", rates_df.isnull().sum())
print("Prices missing values:", prices_df.isnull().sum())

# Check city counts
print("Prices city counts:", prices_df['City'].value_counts())

# Check price ranges
print("Prices price range:", prices_df['Price'].min(), "→", prices_df['Price'].max())

# Summary
prices_df.isna().sum()
rates_df.isna().sum()
```
